```python
import pandas as pd

url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
data_bc = pd.read_csv(url1)
display("Data Set 0")
display(data_bc)

url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
data_bc1 = pd.read_csv(url2)
display("Data Set 1")
display(data_bc1)

url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
data_bc2 = pd.read_csv(url3)
display("Data Set 2")
display(data_bc2)
```

```python
# Concatenate all three datasets into a new DataFrame
data_bc3 = pd.concat([data_bc, data_bc1, data_bc2], ignore_index=True)

# Display the new dataset
print("Data Set 3:")
display(data_bc3)
```

```python
# Print the original column names
print("Original column names:")
print(data_bc3.columns)

# Create a dictionary for replacements
replacement_dict = {
    "Customer": "customer",
    "ST": "state",
    "GENDER": "gender",
    "Education": "education",
    "Customer Lifetime Value": "customer_lifetime_value",
    "Income":"income",
    "Monthly Premium Auto": "monthly_premium_auto",
    "Number of Open Complaints":"number_of_open_complaints",
    "Policy Type": "policy_type",
    "Vehicle Class": "vehicle_class",
    "Total Claim Amount": "total_claim_amount",
    
}

# Rename the columns using the replacement dictionary
data_bc3.rename(columns=replacement_dict, inplace=True)

# Print the first few rows of the updated dataframe
print("Updated dataframe with standardized column names:")
data_bc3
```

```python
# Step 1: Standardize the values in the "Gender" column
data_bc3['gender'] = data_bc3['gender'].str.lower().replace({'femal': 'f', 'male': 'm'})

# Step 2: Replace state abbreviations with full names in the "State" column
state_mapping = {
    'AZ': 'Arizona',
    'Cali': 'California',
    'WA': 'Washington'
}
data_bc3['state'] = data_bc3['state'].replace(state_mapping)

# Step 3: Standardize education levels in the "Education" column
data_bc3['education'] = data_bc3['education'].replace({'Bachelors': 'Bachelor'})

# Step 4: Remove the "%" character from values in the "Customer Lifetime Value" column
data_bc3['customer_lifetime_value'] = data_bc3['customer_lifetime_value'].str.replace('%', '')

# Step 5: Replace certain vehicle classes with a more general category in the "Vehicle Class" column
vehicle_mapping = {
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
}
data_bc3['vehicle_class'] = data_bc3['vehicle_class'].replace(vehicle_mapping)

# Print the first few rows of the cleaned dataframe to verify the changes
print("Cleaned dataframe:")
display(data_bc3)
```

```python
# Ensure "Customer Lifetime Value" is numeric
data_bc3['customer_lifetime_value'] = pd.to_numeric(data_bc3['customer_lifetime_value'], errors='coerce')

# Step 2: Correct the format of "Number of Open Complaints" column and convert it to numeric type
# Convert to string first
data_bc3['number_of_open_complaints'] = data_bc3['number_of_open_complaints'].astype(str)

# Split the string and take the second part
data_bc3['number_of_open_complaints'] = data_bc3['number_of_open_complaints'].str.split('/').str[1]

# Convert to numeric type
data_bc3['number_of_open_complaints'] = pd.to_numeric(data_bc3['number_of_open_complaints'], errors='coerce')

# Print the data types of all columns to verify the changes
print("Updated data types:")
print(data_bc3.dtypes)

display(data_bc3)
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Extract month from the 'effective_to_date	' column
marketing['effective_to_date'] = pd.to_datetime(marketing['effective_to_date'])
marketing['month'] = marketing['effective_to_date'].dt.month

# Create a summary table showing the number of open complaints by policy type and month
complaints_summary = marketing.groupby(['policy_type', 'month'])['number_of_open_complaints'].sum().reset_index()

# Rename the column to match the desired output
complaints_summary.rename(columns={'number_of_open_complaints': 'number_of_complaints'}, inplace=True)

# Display the summary table
display(complaints_summary)
```
